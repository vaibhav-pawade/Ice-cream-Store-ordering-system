# Below Zero - ice cream store
The local ice-cream store needs a new ordering system to improve customer service by streamlining the ordering process. The manager of the store has found that many orders are incorrect and staff at times enter incorrect prices. The manager wants a new ordering system to avoid these errors and to ensure correct prices are charged.
At present, the customer selects what to put the ice-cream in: plain cone, waffle, cone or a cup. The customer also chooses how many scoops and the flavours of each scoop. There are currently eight flavours to choose from.
The management are hoping to have a system developed where staff enters the order at stations where the price is calculated, and the customer pays. The order is then sent to staff at the counters who complete the order and give it to the customer.

## Ordering System
In this first part of Task 2, you should write a Python script for the ice cream ordering system. Your program must store the prices of the individual items (cones, scoop flavours and toppings). Specifically, the program should do the following:
- At the start, the program should ask for the customerâ€™s name to personalise and track the order and then get the number of ice creams from the customer.
- For each ice cream, the cone type, scoop amount, scoop flavour, number and type of toppings should be taken from the customer.
- At the end the program should display the itemised order with the total price.
- The program should be able to handle invalid entries inputted (such as incorrect cone type, or invalid number of toppings)

```python
def order():
    name=input("Enter customer's name :").upper()
    mob=input("Enter customer's mobile number :")
    t=0
    i=0
    cost=0
    count=1
    
    bill={'CONE TYPE':{'PLAIN CONE':[0,1.5],'WAFFLE CONE':[0,2],'CUP':[0,1]},
     'FLAVOUR':{'VANILLA':[0,0.5],'STRAWBERRY':[0,0.5],'CHOCOLATE':[0,0.5],'CARAMEL':[0,0.5],'MINT':[0,0.5],'RAINBOW':[0,0.5],'COFFEE':[0,0.5],'BUBBLE GUM':[0,0.5]},
      'TOPPING':{'PEANUTS':[0,0.75],'CARAMEL SAUCE':[0,0.5],'RAINBOW SPRINKLES':[0,0.5],'PECAN':[0,1],'CHOCOLATE SPRINKLES':[0,0.5]}}
    
    cone_type=['PLAIN CONE','WAFFLE CONE','CUP']
    flavour=['VANILLA','STRAWBERRY','CHOCOLATE','CARAMEL','MINT','RAINBOW','COFFEE','BUBBLE GUM']
    topping=['PEANUTS','CARAMEL SAUCE','RAINBOW SPRINKLES','PECAN','CHOCOLATE SPRINKLES']
  
    
    while(True):
        t=0
        i=0
        print('-------------------------------------------------------------')
        print("Enter "+str(count)+" Ice-Cream details :")
    
        while(True):
            p=input("Enter Cone Type ( Plain Cone / Waffle Cone / Cup ) :").upper()
            if p in cone_type:
                if p=='PLAIN CONE':
                    bill['CONE TYPE']['PLAIN CONE'][0]+=1
                    break
                    
                elif p=='WAFFLE CONE':
                    bill['CONE TYPE']['WAFFLE CONE'][0]+=1
                    break
                    
                elif p=='CUP':
                    bill['CONE TYPE']['CUP'][0]+=1
                    break
                
            else:
                print("!!! Enter Valid Cone type !!!")
            
            
        while(True):
            q=input("Enter Flavour ( Vanilla / Strawberry / Chocolate / Caramel / Mint / Rainbow / Coffee / Bubble gum ) :").upper()
            
            if q in flavour:
                if q=='VANILLA':
                    bill['FLAVOUR']['VANILLA'][0]+=1
            
                elif q=='STRAWBERRY':
                    bill['FLAVOUR']['STRAWBERRY'][0]+=1
                     
                elif q=='CHOCOLATE':
                    bill['FLAVOUR']['CHOCOLATE'][0]+=1
                     
                elif q=='CARAMEL':
                    bill[ 'FLAVOUR']['CARAMEL'][0]+=1
                                     
                elif q=='MINT':
                    bill['FLAVOUR']['MINT'][0]+=1
                                         
                elif q=='RAINBOW':
                    bill['FLAVOUR']['RAINBOW'][0]+=1
                     
                elif q=='COFFEE':
                    bill['FLAVOUR']['COFFEE'][0]+=1
                                         
                elif q=='BUBBLE GUM':
                    bill['FLAVOUR']['BUBBLE GUM'][0]+=1
                                         
                t+=1
                if t==3:
                    print(" Now, you Can't Add more Flavour !!!")
                    break
                else:
                    r=input("Want to Add another Flavour (Y / N)").upper()
                    if r=='N':
                        break 
            else:
                print("!!! Enter Valid Flavour type !!!")
                
                
        while(True):
            o=input("Enter Topping  ( Peanuts /Caramel Sauce / Rainbow Sprinkles / Pecan / Chocolate Sprinkles ) :").upper()
            
            if o in topping:
                if o=='PEANUTS':
                    bill['TOPPING']['PEANUTS'][0]+=1
                     
                elif o=='CARAMEL SAUCE':
                    bill['TOPPING']['CARAMEL SAUCE'][0]+=1
                     
                elif o=='RAINBOW SPRINKLES':
                    bill['TOPPING']['RAINBOW SPRINKLES'][0]+=1
                     
                elif o=='PECAN':
                    bill['TOPPING']['PECAN'][0]+=1
                                     
                elif o=='CHOCOLATE SPRINKLES':
                    bill['TOPPING']['CHOCOLATE SPRINKLES'][0]+=1
                                     
                i+=1
                if i==4:
                    print(" Now, you Can't Add more Topping !!!")
                    break
                else:
                    s=input("Want to Add another Topping (Y / N) :").upper()
                    if s=='N':
                        break 
            else:
                print("!!! Enter Valid Topping type !!!")
        
        z=input("Want to buy more ice-cream  (Y / N) :").upper()
        if z=='N':
            print('-------------------------------------------------------------')
            print('\n')
            print("BELOW ZERO - ICE CREAM STORE",end="\n")
            print("Manish Nagar, Nagpur",end="\n")
            print(".............................................................")
            print('\n')
            print('CUSTOMER\'S NAME : '+name)
            print('CUSTOMER\'S MOBILE NUMBER : '+mob)
            print('*************************************************************')
            for f in bill:
                for g in bill[f]:
                    if bill[f][g][0]>0:
                        print(f+" - "+g+"      $"+str(bill[f][g][1])+" X "+str(bill[f][g][0])+" : $"+str((bill[f][g][1])*(bill[f][g][0])))
                        cost+=(bill[f][g][1])*(bill[f][g][0])
            print('Ice-Cream x '+str(count))
            
            print('*************************************************************')
            print("Amount                $"+str(cost))
            print('-------------------------------------------------------------')
            break 
        elif z=='Y':
            count+=1
```
